# Indexación de Strings

- Indexación significa acceder a los elementos
- Para acceder a los elementos podemos usar corchetes "[]"
- El índice o número que pongamos se refiere a la posición dentro de la lista ordenada
- La indexación puede usarse para extraer elementos individuales de una estructura de datos
- En python, toda la indexación parte desde el número 0 como primera posición
- En Python, tenemos dos tipos de índices:
    - Positivos: comienzan desde 0; estos inidices buscan de izquierda a derecha la posición solicitada
    - Negativos: comienzan desde -1; estos índices buscan de derecha a izquierda la posición solicitada

## Segmentación de strings

```python
# índices positivos
my_str = "Curso de Backend Pyhton"
print(my_str)
print(type(my_str))
print(my_str[0]) # 0
print(my_str[0:5]) # 0,1,2,3,4
print(my_str[:5]) # 0,1,2,3,4
print(my_str[:]) #todos
print(my_str[6:11]) # 6,7,8,9,10
```

```python
#ERROR
my_str = "Curso de Backend Pyhton"
print(my_str[9999])
```

```python
# Índices negativos
```

```python
my_str = "Curso de Backend Pyhton"
print(my_str[len(my_str)-1])
print(my_str[-1])
print(my_str[-5:])
```

## Concatenación de strings

```python
my_str_1 = "Curso"
my_str_2 = "Backen Python"
print(my_str_1)
print(my_str_2)
print(my_str_1+my_str_2)
print(my_str_1+" "+my_str_2)
print("{} {}".format(my_str_1, my_str_2)) #Forma mas corrrecta, mas prodesional
```

```python
my_str_1 = "10"
my_str_2 = "20"
print(my_str_1+my_str_2)
print(eval(my_str_1+my_str_2))
print(eval("{}+{}".format(my_str_1,my_str_2)))
print(int(my_str_1)+int(my_str_2))
```

## Funciones de Strings

### split
- Divide la cadena basándose en el separador especificado

```python
my_str = "El Curso de Backend Python de EBAC te lleva de la mano para que aprendas a programar"
split_result = my_str.split(" ")
print(split_result)
```

```python
my_str = "user3418@hotmail.com"
username, domain = my_str.split("@")
print(username)
print(domain)
```

```python
my_str = """ 2026-01-29|error|Este es un mensaje de error
2026-01-26|warning|Este es un mensaje de advertencia
2026-01-25|info|Este es un mensaje informativo"""
logs = my_str.split("\n")
for log in logs:
    date, log_level, message = log.split("|")
    print("date: ", date)
    print("log_level: ", log_level)
    print("message: ", message)
```

### index/find
-Busca la primer posición de la subcadena solicitada en la cadena deseada

```python
my_str = "El Curso de Backend Python de EBAC te lleva de la mano para que aprendas a programar"
print("La primera posición de 'Pyhton' en my_str es: ", my_str.index("Python"))
print("La primera posición de 'Pyhton' en my_str es: ", my_str.find("Python"))
```

```python
my_str = "El Curso de Backend Python de EBAC te lleva de la mano para que aprendas a programar"
print("La primera posición de 'Backend' en my_str es: ", my_str.index("Backend"))
print("La primera posición de 'Backend' en my_str es: ", my_str.find("Backend"))
```

### Formateo de cadenas y datetime

```python
from datetime import datetime
now = datetime.now()
print(now)
```

```python
year = now.strftime("%Y")
month = now.strftime("%m")
day = now.strftime("%d")
time = now.strftime("%H:%M:%S")
print("Año: ", year)
print("Mes: ", month)
print("Día: ", day)
print("Hora: ", time)
```

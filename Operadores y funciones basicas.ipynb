## Operadores y Tipos de Operadores
¿Qué son los operadores?
- Son un conjunto de símbolos para un lenguaje de programación
- Pueden decirle al compilador o intérprete que ejecute ciertas acciones para producir un resultado final
- Pueden ser usados para aplicar operaciones en variables o valores
- Pueden ser clasificados según su tipo (aritméticos, relacionales, de asignación, lógicos, afiliación y de identidad)

#### Tipos de Operadores:
- Aritméticos
- Relacionales
- De Asignación
- Lógicos
- DE Afiliación
- De Identidad

### Operadores Aritméticos
- "+" : suma
- "-" : resta
- "*" : multiplicación
- "/" : división
- "//" : división con redondeo hacia abajo
- "**" : exponente

```python
# "+" : suma
num_1 = 500
num_2 = 300
print("El resultado de num_1+num_2 es: ",num_1+num_2)
```

```python
# "-" : resta
num_1 = 500
num_2 = 300
print("El resultado de num_1-num_2 es: ",num_1-num_2)
```

```python
# "*" : multiplicación
num_1 = 500
num_2 = 300
print("El resultado de num_1*num_2 es: ",num_1*num_2)
```

```python
# "/" : división
num_1 = 500
num_2 = 300
print("El resultado de num_1/num_2 es: ",num_1/num_2)
```

```python
# "//" : división con redondeo hacia abajo
num_1 = 500
num_2 = 300
print("El resultado de num_1//num_2 es: ",num_1//num_2)
```

```python
# "**" : exponencial
num_1 = 10
num_2 = 5
print("El resultado de num_1**num_2 es: ",num_1**num_2)
```

## Operadores de Asignación

- En Python son usados para asignar valores a variables
- Incluyen al operador de asignacón que es el signo de igual (=)

Tipos de Operadores de Asignación:
- "=" : Operador de Asignación
- "+=" : Sumar y Asignar
- "-=" : Restar y Asignar
- "*=" : Multiplicar y Asignar
- "/=" : Dividir y Asignar
- "//=" : Dividir con redondeo hacia abajo y asignar
- "%=" : Modulo y Asignar
- "**=" : Exponente y asignar

```python
# "=" : Operador de Asignación
```

```python
num_1 = 20
num_2 = 3
```

```python
#"+=" : Sumar y Asignar
```

```python
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 += num_2 #num_1 = num_1 + num_2
print("num_1 después de la operación: ", num_1)
```

```python
# "-=" : Restar y Asignar
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 -= num_2 #num_1 = num_1 - num_2
print("num_1 después de la operación: ", num_1)
```

```python
# "*=" : Multiplicar y Asignar
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 *= num_2 #num_1 = num_1 * num_2
print("num_1 después de la operación: ", num_1)
```

```python
# "/=" : Dividir y Asignar
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 /= num_2 #num_1 = num_1 / num_2
print("num_1 después de la operación: ", num_1)
```

```python
##"//=" : Dividir con redondeo hacia abajo y asignar
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 //= num_2 #num_1 = num_1 // num_2
print("num_1 después de la operación: ", num_1)
```

```python
# "%=" : Modulo y Asignar - El número que queda después de haber dividido (residuo)
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 %= num_2 #num_1 = num_1 % num_2
print("num_1 después de la operación: ", num_1)
```

```python
# "**=" : Exponente y asignar
num_1 = 20
num_2 = 3
print("num_1 antes de la operación: ", num_1)
num_1 **= num_2 #num_1 = num_1 ** num_2
print("num_1 después de la operación: ", num_1)
```

## Operadores lógicos
- Son usados princpialmente para declaraciones condicionales
- Son indicadores estructurales
- También se les conoce como conectores lógicos
- Hay 3 tipos (AND, OR, NOT)
    - AND (y) - regresa True/Verdadero si ambas operaciones son verdaderas
    - OR (o) - regresa True/Verdadero si por lo menos una de las operqciones es verdadera
    - NOT (lo opuesto) - regresa True/Verdadero si la operación es False/Falso

```python
# AND (y)
# Todas las condiciones deben de ser Verdaderas
```

```python
num_1 = 11
num_2 = 55

print("num_1<=num_2 and num_1!=num_2: ",num_1<=num_2 and num_1!=num_2 )
```

```python
num_1 = 11
num_2 = 11

print("num_1<=num_2 and num_1!=num_2: ",num_1<=num_2 and num_1!=num_2 )
```

```python
# OR (o)
# Con que una de las condiciones se cumpla regresa verdadero
```

```python
num_1 = 11
num_2 = 55

print("num_1<=num_2 or num_1!=num_2: ",num_1<=num_2 or num_1!=num_2 )
```

```python
num_1 = 11
num_2 = 11

print("num_1<=num_2 or num_1!=num_2: ",num_1<=num_2 or num_1!=num_2 )
```

```python
# NOT (lo opuesto)
# Si la condición es True regresa False, y si la condición es False regresa True
```

```python
num_1 = 11
num_2 = 55

print("not num_1<=num_2: ", not num_1<=num_2)
```

```python
num_1 = 11
num_2 = 11

print("not num_1!=num_2: ", not num_1!=num_2)
```

## Operadores de Afiliación
- Son usados para validar los elementos en listas, cadenas y tuplas
- Validan si una variable está presente en una secuencia
- Regresan un valor booleano con el resultado (True/False)
- Regresan True si el valor está presente en la secuencia
- Regresan False si el valor no está presente en la secuencia
- Hay 2 operadores de afiliación
    - IN - El valor está en la secuencia
    - NOT IN - El valor no está en la secuencia

```python
# IN: ¿El valor está presente en la secuecnia?
```

## Operadores de Identidad
- Son usados para comparar la ubicación en memoria de dos objetos
- Se utilizan para validar si los valores comparten el mismo espacio en memoria o no
- Regresa True si lo comparten, y False si no lo comparten
- Hay 2 operadores de identidad
    - IS
    - IS NOT

```python
# IS
```

```python
num_1 = 11
num_2 = 55
print("num_1 is num_2: ", num_1 is num_2)
```

```python
num_1 = 77
num_2 = 77
print("num_1 is num_2: ", num_1 is num_2)
```

```python
# IS NOT
```

```python
num_1 = 11
num_2 = 55
print("num_1 is not num_2: ", num_1 is not num_2)
```

```python
num_1 = 77
num_2 = 77
print("num_1 is not num_2: ", num_1 is not num_2)
```

## Funciones Básicas

### trunc()

```python
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("num_1: ", num_1)
print("num_2: ", num_2)
print("num_3: ", num_3)
print("num_4: ", num_4)
```

```python
from math import trunc
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("trunc(num_1): ", trunc(num_1))
print("trunc(num_2): ", trunc(num_2))
print("trunc(num_3): ", trunc(num_3))
print("trunc(num_4): ", trunc(num_4))
```

### round(), ceil() y floor()

```python
# round
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("round(num_1): ", round(num_1))
print("round(num_2): ", round(num_2))
print("round(num_3): ", round(num_3))
print("round(num_4): ", round(num_4))
```

```python
#ceil - redondeo hacia arriba
from math import ceil
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("ceil(num_1): ", ceil(num_1))
print("ceil(num_2): ", ceil(num_2))
print("ceil(num_3): ", ceil(num_3))
print("ceil(num_4): ", ceil(num_4))
```

```python
# floor - redondeo hacia abajo
from math import floor
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("floor(num_1): ", floor(num_1))
print("floor(num_2): ", floor(num_2))
print("floor(num_3): ", floor(num_3))
print("floor(num_4): ", floor(num_4))
```

## Función %

```python
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("'%.1f' % num_1: ", '%.1f' % num_1)
print("'%.3f' % num_2: ", '%.3f' % num_2)
print("'%.4f' % num_3: ", '%.4f' % num_3)
print("'%.5f' % num_4: ", '%.5f' % num_4)
```

## format()

```python
num_1 = 64.99
num_2 = 64.5
num_3 = 64.51
num_4 = 64.00000000001
print("'{:.1f}'.format(num_1): ", '{:.1f}'.format(num_1))
print("'{:.3f}'.format(num_2): ", '{:.3f}'.format(num_2))
print("'{:.4f}'.format(num_3): ", '{:.4f}'.format(num_3))
print("'{:.5f}'.format(num_4): ", '{:.5f}'.format(num_4))
```
